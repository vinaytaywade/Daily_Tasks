```python
import pandas as pd
import numpy as np
```

```python
biden_data=pd.read_csv(r'/content/Bidenall2.csv')
trump_data=pd.read_csv(r'/content/Trumpall2.csv')
```

```python
biden_data.head()
```

```python
trump_data.head()
```

```python
!pip install textblob
```

```python
from textblob import TextBlob
trump_review=TextBlob(trump_data['text'][10])
trump_review.sentiment
```

```python
def find_pol(review):
  return TextBlob(review).sentiment.polarity
```

```python
trump_data['Sentiment Polarity']=trump_data['text'].apply(find_pol)
biden_data['Sentiment Polarity']=biden_data['text'].apply(find_pol)
```

```python
trump_data.head()
```

```python
biden_data.head()
```

```python
trump_data['Expression Label']=np.where(trump_data['Sentiment Polarity']>0,'positive','negative')
```

```python
biden_data['Expression Label']=np.where(biden_data['Sentiment Polarity']>0,'positive','negative')
```

```python
trump_data['Expression Label']=np.where(trump_data['Sentiment Polarity']==0,'neutral',trump_data['Expression Label'])
```

```python
trump_data.head()
```

```python
biden_data['Expression Label']=np.where(biden_data['Sentiment Polarity']==0,'neutral',biden_data['Expression Label'])
```

```python
biden_data.head()
```

```python
filter=trump_data['Expression Label']!='neutral'
tdata=trump_data[filter]
```

```python
tdata.head()
```

```python
tdata.shape
```

```python
filter=biden_data['Expression Label']!='neutral'
bdata=biden_data[filter]
```

```python
bdata.head()
```

```python
bdata.shape
```

```python
tdata.groupby('Expression Label').count()
```

```python
tdata['Expression Label'].value_counts()
```

```python
positive_trump=(tdata['Expression Label'].value_counts()[0]/1324)*100
negative_trump=(tdata['Expression Label'].value_counts()[1]/1324)*100
```

```python
bdata['Expression Label'].value_counts()
```

```python
positive_biden=(bdata['Expression Label'].value_counts()[0]/1031)*100
negative_biden=(bdata['Expression Label'].value_counts()[0]/1031)*100
```

```python
politicians=['Donal Trump','Joe Biden']
listpos=[positive_trump,positive_biden]
listneg=[negative_trump,negative_biden]
```

```python
import plotly
from plotly import graph_objs as go
```

```python
fig=go.Figure(data=[go.Bar(name='positive',x=politicians,y=listpos),go.Bar(name='negative',x=politicians,y=listneg)])
```

```python
fig.show()
```
